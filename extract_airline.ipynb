```python
import xmltodict
import requests
import pandas as pd
```

```python
response = requests.get("https://flydata.avinor.no/airlineNames.asp")
airline_dict = xmltodict.parse(response.content)
```

```python
airline_dict = airline_dict["airlineNames"]["airlineName"]
```

```python
df_original = pd.DataFrame(airline_dict)
df_original.rename(columns={"@code": "code", "@name": "name"}, inplace=True)
df_original
```

```python
df_original.to_json(r"silver\airlines.json", orient="records", indent=2)
```
