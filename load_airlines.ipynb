```python
import psycopg2
import pandas as pd
```

```python
df_original = pd.read_json(r"gold\airlines.json")
df_draft = df_original.copy()
```

```python
with open("db_connection.txt") as db_info:
    conn_string = db_info.readline()

conn = psycopg2.connect(conn_string)
if conn.closed == 0:
    print("Connection established")
```

```python
df_draft
```

```python
cur = conn.cursor()

for index, row in df_draft.iterrows(): 
    formatted_values = [
        f"'{value}'" if isinstance(value, (str, pd.Timestamp)) else str(value)
        for value in row.values
    ]

    query = f"""
        INSERT INTO airlines 
        ({', '.join(df_draft.columns)}) 
        VALUES 
        ({', '.join(formatted_values)});
    """

    try:
        cur.execute(query)
    except Exception as e:
        print(f"Error inserting row {index}: {e}")

cur.close()
```

```python
conn.commit()
```
