```python
with open (filnavn, 'r', encoding='utf-8') as file:
    xml_content= file.read()
    my_ordered_dict=xmltodict.parse(xml_content)
```

```python
clean_xml = my_ordered_dict['airportNames']['airportName']
```

```python

new_list = []

for dictos in clean_xml:
    for key, value in dictos.items():
        new_dict = {}
        new_dict[dictos['@code']] = dictos['@name']
        new_list.append(new_dict)

```

```python
with open("Avinor_Airport_Names.json", "w", encoding='utf-8') as outfile:
    json.dump(new_list, outfile, ensure_ascii=False)
```
