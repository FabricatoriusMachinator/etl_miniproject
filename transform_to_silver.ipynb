```python
import pandas as pd
```

```python
df_flights = pd.read_json(r"bronze\flights.json")
```

```python
df_draft = df_flights.copy()

airlines_to_exclude = ["JUL", "BRR"]

df_draft = df_draft[~df_draft["airline"].isin(airlines_to_exclude)]

df_draft.drop(columns=["gate", "via_airport", "check_in", "status", "delayed"], inplace=True)

df_draft
```

```python
df_draft.rename(columns={"airport": "destination"}, inplace=True)
```

```python
df_draft["schedule_time"] = pd.to_datetime(df_draft["schedule_time"])
df_draft["date"] = df_draft["schedule_time"].dt.date
df_draft["time"] = df_draft["schedule_time"].dt.time
```

```python
df_draft.drop(columns=["schedule_time"], inplace=True)
```

```python
df_draft.to_json(r"silver\flights.json", orient="records", indent=2)
```
